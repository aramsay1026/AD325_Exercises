```python
def max_avg_subarray(nums, k):
    # this is the sliding window technique from the article https://builtin.com/data-science/sliding-window-algorithm
    # starting index of the sliding window
    st = 0
    # keeps track of the largest window sum we've seen so far
    max_sum = float('-inf')
    # running sum of the current window
    win_sum = 0.0

    for i in range(len(nums)):
        # add the current element to the window sum
        win_sum += nums[i]

        # once we've added at least k elements, the window is full
        if i >= k - 1:
            # check if this window has a bigger sum than what we had before
            max_sum = max(win_sum, max_sum)

            # remove the element that's sliding out of the window
            win_sum -= nums[st]

            # move the start of the window to the right by one
            st += 1

    # divide by k to convert the best sum into the best average
    return max_sum / k

# Example usage:
nums = [1, 12, -5, -6, 50, 3]
k = 4
print(max_avg_subarray(nums, k))  # Output: 12.75

nums2 = [5]
k = 1
print(max_avg_subarray(nums2, k))  # Output: 5.0

```
